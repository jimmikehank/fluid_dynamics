```python
def turn_pot(n,A):
    import numpy as np

    Ny = 101
    Nx = 101
    x = np.linspace(-1,1,Nx)
    y = np.linspace(-1,1,Ny)

    F_real = np.zeros((Ny,Nx))
    F_imag = np.zeros((Ny,Nx))

    for i in range(Nx):
        for j in range(Ny):
            z = complex(x[i],y[j])
            if np.real(z) == 0:
                z += complex(0.00001,0)
            if np.imag(z) == 0:
                z += complex(0,0.00001)
            F_real[j,i] = np.real(A*n*z**(n-1))
            F_imag[j,i] = np.imag(A*n*z**(n-1))
    return F_real, F_imag, x, y

def vortex_pot(m,U,g,e):
    import numpy as np

    Ny = 51
    Nx = 51
    x = np.linspace(-2,2,Nx)
    y = np.linspace(-2,2,Ny)

    F_real = np.zeros((Ny,Nx))
    F_imag = np.zeros((Ny,Nx))

    for i in range(Nx):
        for j in range(Ny):
            z = complex(x[i],y[j])
            if np.real(z) == 0:
                z += complex(0.00001,0)
            if np.imag(z) == 0:
                z += complex(0,0.00001)
            
            W = U+(m+m/2*1j)/(np.pi*(2.*(z+e))) - (m-m/2*1j)/(np.pi*2.*(z-e))
            F_real[j,i] = np.real(W)
            F_imag[j,i] = np.imag(W)
    return F_real, F_imag, x, y, W
```

```python
import numpy as np
Fr,Fi,x,y, W = vortex_pot(100,1,0.5,1)
X,Y = np.meshgrid(x,y)
import matplotlib.pyplot as plt

plt.streamplot(x,y,Fr,-Fi)

```
